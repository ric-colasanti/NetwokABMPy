```python
import networkx as nx
import random as rnd
```

```python
g = nx.grid_2d_graph(10,10)
```

```python
position =list(g.nodes)
```

```python
pos = {point: point for point in position}
```

```python
nx.draw(g,pos=pos)
```

```python
nx.set_node_attributes(g,0,"val")
c=0
for y in range(10):
    for x in range(10):
        g.nodes[x][y]["val"]=y*10+x
        c+=1
    
```

```python
g.nodes[(1,1)]
```

```python
shortest_path = nx.shortest_path(g, (1,1), (3,4))  # Specify weight attribute
```
